```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Asumsikan df sudah disiapkan dan lengkap 1 jam 1 row
# Kolom: Date, Hour, Order

# --- PREPARE DATA ---
df = df.sort_values(by=["Date", "Hour"])
orders = df["Order"].values.astype(float)

# Normalisasi
scaler = MinMaxScaler()
orders_scaled = scaler.fit_transform(orders.reshape(-1, 1))

# Buat sequence: 168 jam input → 24 jam output
X, y = [], []
window_in, window_out = 168, 24

for i in range(len(orders_scaled) - window_in - window_out):
    X.append(orders_scaled[i : i + window_in])
    y.append(orders_scaled[i + window_in : i + window_in + window_out])

X = np.array(X)
y = np.array(y)

# Reshape for LSTM: (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# --- LSTM MODEL ---
model = Sequential()
model.add(LSTM(64, input_shape=(window_in, 1)))
model.add(Dense(window_out, activation="sigmoid"))
model.compile(optimizer="adam", loss="mse")

# --- TRAINING ---
model.fit(X, y, epochs=10, batch_size=16)

# --- PREDIKSI BESOK ---
latest_input = orders_scaled[-window_in:].reshape((1, window_in, 1))
pred_scaled = model.predict(latest_input)[0]
pred = scaler.inverse_transform(pred_scaled.reshape(-1, 1)).flatten()

# --- CEK JAM YANG DIPREDIKSI ADA ORDER ---
threshold = 0.5  # bisa disesuaikan
pred_order_hours = [i for i, val in enumerate(pred) if val > threshold]

print("Jam yang diprediksi ada order besok:", pred_order_hours)
```
